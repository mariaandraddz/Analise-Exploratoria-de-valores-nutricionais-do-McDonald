```python
#Criando um dicionário com as traduções
novas_colunas = {
    'Menu Category':'Categoria',
    'Menu Items': 'Itens de menu',
    'Per Serve Size': 'Por tamanho de porção',
    'Energy (kCal)': 'Energia (kcal)',
    'Protein (g)' :'Proteína (g)',
    'Total fat (g)': 'Gordura Total',
    'Sat Fat (g)': 'Gordura Saturada (g)',
    'Trans fat (g)': 'Gordura Trans (g)',
    'Cholesterols (mg)': 'Colesteróis (mg)',
    'Total carbohydrate (g)': 'Carboidratos totais (g)',
    'Total Sugars (g)': 'Açúcares Totais (g)',
    'Added Sugars (g)': 'Açúcares adicionados (g)',
    'Sodium (mg)': 'Sódio (mg)'
}

dataframe = dataframe.rename(columns=novas_colunas)
dataframe.columns
```

# TRATAMENTO DE DADOS

```python
#Verificar a existência de valores nulos. False para valor existente e True para nulo
dataframe.isnull()
```

```python
#Verificar qual é a soma de linhas que não estão preenchidaa
dataframe.isnull().sum()
```

Foi possível identificar que tem um valor não existente na coluna Sódio (mg)

```python
dataframe[dataframe.isnull().any(axis=1)] #Para identificar qual linha está vazia
```

Identifiquei valores ausentes na coluna de Sódio.
Optei por preencher com a mediana da categoria para evitar que outliers (itens extremamente salgados) distorcessem a média do grupo.

```python
dataframe.head(50)
```

```python
# Preenchendo valores nulos com a média da categoria correspondente
dataframe['Sódio (mg)'] = dataframe['Sódio (mg)'].fillna(dataframe.groupby('Categoria')['Sódio (mg)'].transform('mean'))
```

```python
dataframe[dataframe.isnull().any(axis=1)] #Para identificar qual linha está vazia
```

É possível verificar que a linha vazia foi preenchida

# EXPLORAÇÃO E VISUALIZAÇÃO DE DADOS

Qual é o item mais calórico de todo o cardápio? E o menos calórico?

```python
dataframe['Categoria'].value_counts().plot(kind='bar', title='Itens Caloricos')
```

```python
#Para calcular a quantidade de itens por categoria
itens = dataframe['Categoria'].value_counts()
```

```python
plt.figure(figsize=(15, 6)) #Tamanho da figura
sns.barplot(x=itens.index, y= itens.values, palette='Blues', hue=itens.values)
plt.title('Itens')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.show()
```

Percebemos que "McCafé Menu" é a categoria com maior quantidade

Quais itens contem a maior quantidade de sódio?

```python
ordem = dataframe.groupby('Categoria')['Sódio (mg)'].mean().sort_values(ascending=False).index
print(ordem)
```

```python
plt.figure(figsize=(15, 4)) #tamanho da figura
sns.barplot(data=dataframe, x='Categoria', y='Sódio (mg)', errorbar=None, palette = 'Blues', order=ordem)
plt.title('Itens com maior quantidade de sódio')
plt.xlabel('Categoria')
plt.ylabel('Sódio (mg)')
plt.show()
```
